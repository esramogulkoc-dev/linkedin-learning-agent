```python
# Imports
from dotenv import load_dotenv
import os
import requests
from google import genai

# Load environment variables
load_dotenv()
GOOGLE_API_KEY = os.getenv("GOOGLE_API_KEY")
LINKEDIN_ACCESS_TOKEN = os.getenv("LINKEDIN_ACCESS_TOKEN")
LINKEDIN_PERSON_URN = os.getenv("LINKEDIN_PERSON_URN")

# Initialize Google GenAI client
client = genai.Client(api_key=GOOGLE_API_KEY)

# --- Step 1: Generate post content ---
user_input = """
This week I learned about word embeddings and similarity metrics in NLP.
I also started experimenting with Google Gemini API.
"""
google_model = "models/gemini-2.5-flash-lite"

response = client.models.generate_content(
    model=google_model,
    contents=f"Turn this text into a professional LinkedIn post:\n{user_input}"
)

post_content = response.text
print("Generated Post:\n", post_content)

# --- Step 2: Post to LinkedIn ---
linkedin_url = "https://api.linkedin.com/v2/ugcPosts"
headers = {
    "Authorization": f"Bearer {LINKEDIN_ACCESS_TOKEN}",
    "Content-Type": "application/json",
    "X-Restli-Protocol-Version": "2.0.0"
}

payload = {
    "author": LINKEDIN_PERSON_URN,
    "lifecycleState": "PUBLISHED",
    "specificContent": {
        "com.linkedin.ugc.ShareContent": {
            "shareCommentary": {
                "text": post_content
            },
            "shareMediaCategory": "NONE"
        }
    },
    "visibility": {"com.linkedin.ugc.MemberNetworkVisibility": "PUBLIC"}
}

# Send the POST request
response = requests.post(linkedin_url, headers=headers, json=payload)
if response.status_code == 201:
    print("✅ Post successfully published on LinkedIn!")
else:
    print("❌ Failed to post:", response.status_code, response.text)
```

```python
from google import genai
import os
from dotenv import load_dotenv

# Load environment variables
load_dotenv()
API_KEY = os.getenv("GOOGLE_API_KEY")
client = genai.Client(api_key=API_KEY)

# Your weekly journal input
weekly_journal = """
This week in the Data Analytics Bootcamp, I explored pandas for data cleaning and matplotlib for visualization.
My key insight: Cleaning messy datasets carefully makes analysis way more reliable.
"""

# Prompt for the AI: generate LinkedIn posts (simulation only)
prompt = f"""
You are a LinkedIn post assistant for a data analytics bootcamp student.
1. Read the following weekly journal:
{weekly_journal}

2. Extract the main topic(s) learned this week.
3. Extract the key insight or learning point.
4. Generate 3 LinkedIn post options using these extracted points, in enthusiastic and professional style.
5. Include hashtags like #dataanalytics #bootcamp #learning #dataskills.
6. DO NOT post anything to LinkedIn. Only output the generated posts as text.
"""

# Choose a model
google_chosen_model = "models/gemini-2.5-flash-lite"

# Generate content
response = client.models.generate_content(
    model=google_chosen_model,
    contents=prompt
)

# Save output to file
output_file = "generated_linkedin_posts.txt"
with open(output_file, "w", encoding="utf-8") as f:
    f.write(response.text)

# Print confirmation
print(f"LinkedIn post simulation complete! Posts saved in {output_file}")
```
